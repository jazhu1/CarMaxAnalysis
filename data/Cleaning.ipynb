Cleaning Columns

```python
#price is fine, no missing values already int
# appraisal offer is fine, no misisng values already int
```

```python
#dummy variables for make_appraisal 
df['make_appraisal'].unique()
```

```python
df = pd.get_dummies(df, columns=['make_appraisal'], prefix='', prefix_sep='make_ap_', dtype= int)
```

```python
#dummy variables for model_appraisal
df = pd.get_dummies(df, columns=['model_appraisal'], prefix='', prefix_sep='model_ap_', dtype= int)
```

```python
#creating "not specified" value for nans within trim_level_appraisal, then one hot encoding
df['trim_level_appraisal'] = df['trim_level_appraisal'].fillna('Not Specified')

df = pd.get_dummies(df, columns=['trim_level_appraisal'], prefix='', prefix_sep='TLA_', dtype = int)
```

```python
#model_year_appraisal as int is fine
```

```python
# remove L from engine_appraisal
df['engine_appraisal'] = df['engine_appraisal'].str.rstrip('L')
df['engine_appraisal'] = df['engine_appraisal'].astype(float)
```

```python
#cylinders_appraisal is fine
```

```python
# treating nans in mpg_city_appraisal 
df['mpg_city_appraisal'].unique()
```

```python
#use predictive imputer on mpg_city_appraisal after all columns are cleaned 
```

```python
nan_rows = df[df['mpg_city_appraisal'].isna()]

# Display the rows where 'mpg_city_appraisal' is NaN
print(nan_rows)
```

```python
# mpg_highway_appraisal and horsepower_appraisal are fine
```

```python
#use predictive imputer on fuel_capacity_appraisal after all columns are cleaned 
nan_rows = df[df['fuel_capacity_appraisal'].isna()]

print(nan_rows)
```

```python
#vehicle_type_appraisal, color_appraisal, make, model
selected_columns = ['vehicle_type_appraisal', 'color_appraisal', 'make', 'model']
selected_df = df[selected_columns]
print(selected_df)
```

```python
columns_to_encode = ['vehicle_type_appraisal', 'color_appraisal', 'make', 'model']

# Loop through each column and apply one-hot encoding with prefixes
for column in columns_to_encode:
    df = pd.get_dummies(df, columns=[column], prefix='', prefix_sep=f'{column[:3]}_', dtype=int)
```

```python
#creating "not specified" value for nans within trim_level, then one hot encoding
df['trim_level'] = df['trim_level'].fillna('Not Specified')

df = pd.get_dummies(df, columns=['trim_level'], prefix='', prefix_sep='TL_', dtype = int)
```

```python
#model_year, mileage are fine
```

```python
#renaming engine to include units and separating string
df = df.rename(columns={'engine': 'engine_L'})
df['engine_L'] = df['engine_L'].str.replace('L', '')
```

```python
#cylinders is fine
```

```python
#clean mpg_city and highway with imputing after other rows are cleaned
```

```python
#horsepower, fuel_capacity are fine
```

```python
#dummy variable for color and vehicle type
df = pd.get_dummies(df, columns=['vehicle_type'], prefix='', prefix_sep='VT_', dtype= int)
df = pd.get_dummies(df, columns=['color'], prefix='', prefix_sep='V_', dtype= int)
```

```python
#cleaning online_appraisal_flag, state, days_since_offer
```

```python
#Converting online_appraisal_flag to cat and getting dummies
df['online_appraisal_flag'] = df['online_appraisal_flag'].map({1: 'Online', 0: 'In Person'}).fillna('Not Specified')
df = pd.get_dummies(df, columns=['online_appraisal_flag'], prefix='App_Flag', dtype=int)
```

```python
#days_since_offer is fine
```

```python
#imputing mpg_city_appraisal, fuel_capacity_appraisal, mpg_city, mpg_highway
columns_with_nan = df.columns[df.isna().any()].tolist()

# Print the columns with NaN values
print(columns_with_nan)
```
